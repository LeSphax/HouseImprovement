```python
import csv
import matplotlib.pyplot as plt
import numpy as np

time_labels = []
time = []
power = []
with open('data.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=';', quotechar='|')
    i = -1
    start = 1
    print('Reading...')
    for row in spamreader:
        i+=1
        if i < start: continue
        if row[2] == '?': continue
        #if i % 2 == 0:
        time_labels.append(row[1][0:5])
        time.append(i-start)
        power.append(float(row[2]))
        if i > start + 5000: break
```

```python
print('Plotting...')
plt.plot(time, power)
plt.xlabel('Time')
plt.ylabel('Power')
plt.ylim([0, 7])
indices = np.arange(0, len(time_labels) -1, (len(time_labels) - 1)//7)
print(indices)
plt.xticks(ticks=np.arange(8) * (len(time_labels) - 1)//7, labels=np.take(time_labels, indices))
plt.show()
```
